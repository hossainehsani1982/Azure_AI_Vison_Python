```python
import torch
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    BitsAndBytesConfig,
    DataCollatorForLanguageModeling,
    EarlyStoppingCallback,
)
from datasets import load_dataset
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
import evaluate

# Enable CUDA optimizations
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.benchmark = True
torch.backends.cudnn.deterministic = False

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Enable 4-bit quantization to save memory
from transformers import BitsAndBytesConfig

# Enable 4-bit quantization to save memory
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    llm_int8_threshold=6.0
)


# Load Phi-3 Mini model and tokenizer with quantization
model_name = "microsoft/phi-3-mini-4k-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=quantization_config,
    device_map={"": 0},  # ✅ Force entire model to GPU 0
)

# # Load model in low precision on CPU
# model = AutoModelForCausalLM.from_pretrained(
#     model_name,
#     torch_dtype=torch.bfloat16,  #
#     device_map="cpu"
# )


# Prepare the model for LoRA fine-tuning
#model = prepare_model_for_kbit_training(model, use_reentrant=False)
model = prepare_model_for_kbit_training(model)

# LoRA configuration
lora_config = LoraConfig(
    r=8,  # Smaller rank
    lora_alpha=16,  # Smaller alpha
    lora_dropout=0.1,
    bias="none",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                    "gate_proj", "up_proj", "down_proj"]
)

# Wrap model with LoRA
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# Load datasets
# Stream dataset to avoid loading into RAM all at once
train_dataset = load_dataset("json", data_files="/content/train_dataset.json", streaming=True)["train"]
test_dataset = load_dataset("json", data_files="/content/test_dataset.json", streaming=True)["train"]


# Tokenization function
# def tokenize_function(examples):
#     # Concatenating question and answer as a single text sequence
#     text = [f"Q: {q} A: {a}" for q, a in zip(examples["question"], examples["answer"])]

#     return tokenizer(
#         text,
#         truncation=True,
#         padding="max_length",
#         max_length=512
#     )

# def tokenize_function(examples):
#     tokenized_inputs = tokenizer(
#         examples["question"],
#         text_pair=examples["answer"],
#         truncation=True,
#         padding="max_length",
#         #max_length=512
#         max_length=256

#     )
#     tokenized_inputs["labels"] = tokenized_inputs["input_ids"].copy()
#     return tokenized_inputs


# Tokenize datasets
# tokenized_train = train_dataset.map(tokenize_function, batched=True, batch_size=8)
# tokenized_test = test_dataset.map(tokenize_function, batched=True, batch_size=8)

# Load datasets
train_dataset = load_dataset("json", data_files="/content/train_dataset.json")["train"]
val_dataset = load_dataset("json", data_files="/content/val_dataset.json")["train"]
test_dataset = load_dataset("json", data_files="/content/test_dataset.json")["train"]

# Tokenization function for instruction tuning
def tokenize_function(examples):
    text = [f"User: {q}\nAssistant: {a}" for q, a in zip(examples["prompt"], examples["response"])]

    tokenized_inputs = tokenizer(
        text,
        truncation=True,
        padding="max_length",
        max_length=512
    )
    tokenized_inputs["labels"] = tokenized_inputs["input_ids"].copy()

    return tokenized_inputs

# # Tokenize datasets
tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_val = val_dataset.map(tokenize_function, batched=True)


# Use DataCollator for dynamic padding
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

training_args = TrainingArguments(
    output_dir="./phi3-finetuned",
    evaluation_strategy="epoch",
    learning_rate=3e-5,  # Adjusted learning rate
    per_device_train_batch_size=1,  # Increased batch size
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=4,  # Adjusted accumulation steps
    num_train_epochs=15,  # Increased epochs
    weight_decay=0.01,
    save_strategy="epoch",
    logging_dir="./logs",
    logging_steps=10,
    fp16=True,
    max_grad_norm=1.0,
    #max_grad_norm=0.5
)


# Load accuracy metric
metric = evaluate.load("accuracy")

# Compute accuracy
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = torch.argmax(torch.tensor(logits), dim=-1).cpu().numpy()

    # Compute accuracy
    accuracy = metric.compute(predictions=predictions, references=labels)["accuracy"]

    # Debugging: Print available metrics
    print(f"Available Metrics: {{'accuracy': {accuracy}}}")

    return {"accuracy": accuracy}  # Ensure this matches 'metric_for_best_model'


# Create Trainer instance
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

# Start fine-tuning
trainer.train()

# Save the fine-tuned model
trainer.save_model("./phi3-finetuned")
tokenizer.save_pretrained("./phi3-finetuned")
print("Fine-tuned model saved to ./phi3-finetuned")

# Load the fine-tuned model for testing
model_path = "./phi3-finetuned"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path).to(device)

# Test the model on the test dataset
model.eval()
for example in test_dataset:
    question = example["question"]
    inputs = tokenizer(question, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = model.generate(
            inputs["input_ids"],
            max_length=100,
            num_return_sequences=1,
            temperature=0.7,
            top_k=50,
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"Question: {question}")
    print(f"Generated Response: {response}")
    print("-" * 50)

# Example: Generate a response for a custom question
test_question = "What is the purpose of the High Side image orientation?"
inputs = tokenizer(test_question, return_tensors="pt").to(device)

with torch.no_grad():
    outputs = model.generate(
        inputs["input_ids"],
        max_length=100,
        num_return_sequences=1,
        temperature=0.7,
        top_k=50,
    )

response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"Question: {test_question}")
print(f"Generated Response: {response}")
```
